```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
###calculate the sum of the squared errors(SSE)

SSE = []
for k in range(1, 20):
    estimator = KMeans(n_clusters = k)
    estimator.fit(ingredient_data_norm)
    SSE.append(estimator.inertia_)
X = range(1, 20)
plt.xlabel('k')
plt.ylabel('SSE')
plt.plot(X, SSE, 'o-')
plt.show()
```

```python
###calculate the Silhouette Coefficient(轮廓系数，评估值越大聚类效果越好)

Scores = []  # 
for k in range(2,20):
    estimator = KMeans(n_clusters=k)  # 构造聚类器
    estimator.fit(ingredient_data_norm)
    Scores.append(silhouette_score(ingredient_data_norm,estimator.labels_,metric='euclidean'))
X = range(2,20)
plt.xlabel('k')
plt.ylabel('轮廓系数')
plt.plot(X,Scores,'o-')
plt.show()
```

```python
##to varify the best selection, use following package
from yellowbrick.cluster.elbow import kelbow_visualizer
kelbow_visualizer(KMeans(random_state = 4), ingredient_data_norm, k = (2,20))
```

```python
##KMeans, optimal k = 8
kmeans = KMeans(n_clusters=8)
kmeans.fit(ingredient_data_norm)
cluster_df = ingredient_data_norm.copy()
cluster_df['cluster'] = kmeans.predict(cluster_df)
```

```python
cluster_df
```

```python
col_name = list(cluster_df.columns[0:9])
for c in range(len(col_name)):
    if c == 0:
        cluster_mean = cluster_df.groupby('cluster')[col_name[c]].describe().iloc[:, :2]
        cluster_mean.columns = ['formulation_count', col_name[c]]
    else:
        cluster_mean[col_name[c]] = cluster_df.groupby('cluster')[col_name[c]].describe().iloc[:, 1:2]
cluster_mean
```
